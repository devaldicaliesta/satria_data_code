```python
from sklearn.metrics import accuracy_score
from sklearn.tree import DecisionTreeClassifier
from sklearn.model_selection import GridSearchCV

# Parameter grid for GridSearchCV
param_grid = {
    'criterion': ['gini', 'entropy'],
    'max_depth': [None, 10, 20],
    'min_samples_split': [2, 5, 10],
    'min_samples_leaf': [1, 2, 4]
}

# Decision Tree
dt_model = DecisionTreeClassifier()

# GridSearchCV to find best parameters
grid_search = GridSearchCV(dt_model, param_grid, cv=5, scoring='accuracy')
grid_search.fit(X_train, y_train)

# Get the best model
best_dt_model = grid_search.best_estimator_

# Predict using the best model
dt_pred = best_dt_model.predict(X_test)

# Calculate accuracy
dt_accuracy = accuracy_score(y_test, dt_pred)

# Print best parameters, classification report, and accuracy
print("Best Parameters:", grid_search.best_params_)
dt_classification_report = classification_report(y_test, dt_pred)
print("Decision Tree Classification Report:\n", dt_classification_report)
print("Accuracy:", dt_accuracy)
```
